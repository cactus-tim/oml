Тасочка номер один

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/marketing_campaign.csv', sep=';')
```

```python
def count_complain(df) -> float:
    return df["Complain"].sum()

def avg_buyer_age(df) -> float:
    return 2024 - df[(df["MntFishProducts"] > 0) | (df["MntFruits"] > 0) | (df["MntGoldProds"] > 0) | (df["MntMeatProducts"] > 0) | (df["MntSweetProducts"] > 0) | (df["MntWines"] > 0)]["Year_Birth"].mean()

def avg_age(df) -> float:
    return 2024 - df["Year_Birth"].mean()
```

Подсчет жалоб

```python
count_complain(df)
```

Средний возраст покупателей

```python
avg_buyer_age(df)
```

Средний возраст всех людей в датасете

```python
avg_age(df)
```

Заметим, что средний возраст покупателей равен среднему возрасту всех людей в датасете
Если немного поиграться с самим датасетом, то можно заметить, что все люди в датасете совершали покупки
